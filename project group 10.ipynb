```python
#Group 10 
#Class A3 & B5 with log in system and division to the power of n
from yes_cat import no_rat
def menu():#a text menu as a user interface
    print('Choose a function:\n1) Comparison\n2) Division\n3) Division to the power of n\n4) Exit')
    ch=input('Your choice: ')#ask user to choose a function
    if ch=='1':
        an=int(input('Enter the numerator of a:'))
        ad=int(input('Enter the denominator of a:'))
        bn=int(input('Enter the numerator of b:'))
        bd=int(input('Enter the denominator b:'))
        # enter a rational number and performs a comparison with another rational number
        print(f"Result of subtraction: {no_rat(an, ad).subtract(no_rat(bn, bd))}")
    elif ch=='2':
        a=int(input('Enter the numerator of dividend:'))
        b=int(input('Enter the denominator of dividend:'))
        ad=int(input('Enter the numerator of divisor:'))
        bd=int(input('Enter the denominator divisor:'))
        #enter a rational number and performs a division with another rational number
        print(f"Result of division: {no_rat(a,b).divide(no_rat(ad,bd))}")
        #returns the result as a rational number
    elif ch=='3':
        a=int(input('Enter the numerator:'))
        b=int(input('Enter the denominator:'))
        n=int(input('Enter the power n:'))
        result=no_rat(a,b).power(n)
        #compute a non-zero rational number to the power of n
        if result:
            print(f"Result of division to the power of {n}: {result}")
    elif ch=='4':
        print('Application ended')
        return False
        #The user can also choose to end the application through the interface.
    else:
        print('Invalid choice')
    return True
def main():
    username=input('Enter your username:')
    #login with a username first
    account_exists=False
    f=open('account.txt','r')
    for line in f:
        #the username will be checked with a file that stores username-password pairs
        if line.strip()==username:
            account_exists=True
            break
    f.close()
    if account_exists:
        count=0
        f=open('account.txt','r')
        for line in f:
            if line.strip()==username:
                stored_password=next(f).strip()
                password=input('Enter your password:')
                if password==stored_password:
                    #the program will start only when the password is correct
                    print('Welcome!')
                    while menu():
                        pass
                    return
                else:
                    #The application will end if the user fails to provide the correct password in five consecutive trials.
                    print('Wrong password')
                    count+=1
                    while count<5:
                        password_attempt=input('Try again:')
                        if password_attempt==stored_password:
                            print('Welcome!')
                            while menu():
                                pass
                            return
                        count+=1
                    else:
                        print('You have failed to log in. Application ended')
                        return False
        f.close()
    else: #the username is new
        f=open('account.txt','a')
        print('This account have not sighed up')
        password=input('Enter your password for new account:')
        #ask for a password from the user and store it into that file.
        f.write('\n'+username)
        f.write('\n'+password)
        print('Account created. You can now use the program.')
        while menu():
            pass
if __name__ == "__main__":
    main()
```
